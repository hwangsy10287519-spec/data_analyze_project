```python
import numpy as np
import pandas as pd
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import matplotlib.pyplot as plt

common_cols = df.columns

# 제외할 컬럼
exclude_cols = ['Happiness Score', 'Region', 'Country','Happiness Rank']

feature_cols = [col for col in common_cols if col not in exclude_cols]
print("\n사용할 Feature 컬럼:", feature_cols)

X_train_df = df[feature_cols].drop(156)
X_test_df = df2[feature_cols]

train_data = np.column_stack([X_train_df.index, X_train_df.values])
test_data = np.column_stack([X_test_df.index, X_test_df.values])

y_train = df['Happiness Score'].values
y_test = df2['Happiness Score'].values

y_train_int = np.clip((y_train).astype(int), 0, 9)
y_test_int = np.clip((y_test).astype(int), 0, 9)

train_labels = to_categorical(y_train_int, num_classes=10)
test_labels = to_categorical(y_test_int, num_classes=10)

model = Sequential([
    Input(shape=(train_data.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_data,
    train_labels,
    epochs=200,
    batch_size=128,
    validation_data=(test_data, test_labels)
)

test_loss, test_accuracy = model.evaluate(test_data, test_labels)
print(f"\nTest accuracy: {test_accuracy * 100:.2f}%")
```
